PROBLEM STATEMENT : Using the SMS spam collection dataset to train the data model to filter SMS into SPAM or Important.

```python
import textblob
```

**Importing the Dataset**

```python
df = pd.read_csv('../input/spam.csv', encoding = 'Latin-1') 
df.head()
```

**Deleting the Useless Columns**

```python
df.drop(df.columns[[2,3,4]], axis =1)
```

**Using Vectorizer to clean the text and splitting data into Training and Testing Set**

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
c_vec = CountVectorizer(lowercase=1,min_df=.00001,stop_words='english')
c_vec.fit(df['v2'].values)
X_train = c_vec.transform(df['v2'][0:5000].values)
Y_train = df['v1'][0:5000].values
X_test = c_vec.transform(df['v2'][5000:].values)
Y_test = df['v1'][5000:].values
```

**Fitting the NB classifier**

```python
clf_nb = MultinomialNB()
clf_nb.fit(X_train,Y_train)
clf_nb.score(X_train,Y_train)
```

**Fitting the model to Testing data**

```python
clf_nb.score(X_test,Y_test)
```
